```python
import pandas as pd #import pandas library
data = pd.read_csv("marketing_campaign.csv",sep=None,engine="python") # read the data
```

```python
print("First 5 Data Rows: ")
print(data.head(5)) #sample data set
```

```python
print("Duplicate Values: ")
print(data.isnull().sum())
```

```python
# data = data.dropna() To remove the NaN rows
```

```python
data['Income'] = data['Income'].fillna(data['Income'].mean()) # fill NaN values with mean of income values
```

```python
print(data.isnull().sum())
```

```python
data.drop_duplicates(inplace=True) # drop duplicate rows
```

```python
# replace the column names in lower case and replace white space with underscore
data.columns = data.columns.str.lower().str.replace(" ", "_")
```
